# Renaming columns

To facilitate the movement from the input files to the staging tables, we are going rename some columns of the file.

```python
out = suicide_rates.rename(index=str, columns={'age': 'age_range', 
                                               'suicides/100k pop': 'suicides_100k', 
                                               ' gdp_for_year ($) ': 'gdp_for_year',
                                               'gdp_per_capita ($)': 'gdp_per_capita',
                                               'country-year': 'country_year',
                                               'HDI for year': 'hdi_for_year'
                                              })
out.head(5)
```

```python
out.to_csv("./../datasets/results/suicide_rates.csv")
```
