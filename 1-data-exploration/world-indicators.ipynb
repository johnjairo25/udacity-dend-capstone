# Research Question

Does government expenditure on education (as % of GDP) is related to the IncomeGroup of the countries?

- _IndicatorName_: Government expenditure on education as % of GDP


## Cleaning the data

```python
print('Total number of countries: ', len(countries))

mask = pd.notnull(countries['IncomeGroup'])
cleaned_countries = countries[mask][['CountryCode', 'ShortName', 'IncomeGroup', 'Region']]

print('Countries with IncomeGroup: ', len(cleaned_countries))
```

```python
research_idc = 'Government expenditure on education as % of GDP'
mask = indicators['IndicatorName'].str.contains(research_idc)
indicators_to_research = indicators[mask]
len(indicators_to_research)
```

```python
indicators_to_research.to_csv('indicators_to_research.csv')
```

## Analysis for Colombia

```python
mask = indicators_to_research['CountryCode'].str.contains('COL')
colombia_indicators = indicators_to_research[mask]
colombia_indicators['Year'].values
```

```python
years = colombia_indicators['Year'].values
values = colombia_indicators['Value'].values
plt.plot(years, values)

plt.xlabel('Year')
plt.ylabel(colombia_indicators['IndicatorName'].iloc[0])
plt.title('Colombian Government expenditure on education as % of GDP')

# to make more honest, start they y axis at 0
plt.axis([min(years), max(years), 0, max(values)])

plt.show()
```

## COL, ARG, USA, CAN Analysis

```python
col_ind = indicators_to_research[indicators_to_research['CountryCode'].str.contains('COL')]
arg_ind = indicators_to_research[indicators_to_research['CountryCode'].str.contains('ARG')]
usa_ind = indicators_to_research[indicators_to_research['CountryCode'].str.contains('USA')]
can_ind = indicators_to_research[indicators_to_research['CountryCode'].str.contains('CAN')]

plt.plot(col_ind['Year'].values, col_ind['Value'].values, label = 'Colombia')
plt.plot(arg_ind['Year'].values, arg_ind['Value'].values, label = 'Argentina')
plt.plot(usa_ind['Year'].values, usa_ind['Value'].values, label = 'USA')
plt.plot(can_ind['Year'].values, can_ind['Value'].values, label = 'Canada')

plt.legend(loc = 'upper left')

plt.xlabel('Year')
plt.ylabel('Education Expenditure (%of GDP)')

plt.show()
```

```python
print('Colombia, mean: ', col_ind['Value'].mean(), ', std: ', col_ind['Value'].std())
print('Argentina, mean: ', arg_ind['Value'].mean(), ', std: ', arg_ind['Value'].std())
print('USA, mean: ', usa_ind['Value'].mean(), ', std: ', usa_ind['Value'].std())
print('Canada, mean: ', can_ind['Value'].mean(), ', std: ', can_ind['Value'].std())
```

```python
data = [col_ind['Value'].values, arg_ind['Value'].values, usa_ind['Value'].values, can_ind['Value'].values]
labels = ['COL', 'ARG', 'USA', 'CAN']
plt.boxplot(data, labels = labels)
plt.ylabel('Education Expenditure (%of GDP)')
plt.show()
```

As expected, there seems to exist a relationship between the education expenditure and the income level. USA and Canada have greater income and also have higher education expenditure.

# Analyzing education expenditure with income group

```python
cleaned_countries.head(5)
```

```python
indicators_to_research.head(5)
```

```python
mask = indicators_to_research['Year'] < 2000
old_ind = indicators_to_research[mask]
total_ind = cleaned_countries.merge(old_ind, on = 'CountryCode', how = 'inner')
total_ind.head(5)
```

```python
groups = total_ind['Region'].unique()
groups = list(groups)
groups.sort()
print(groups)
```

```python
groups_dic = {}
for income in groups:
    mask = total_ind['Region'].str.contains(str(income))
    groups_dic[income] = total_ind[mask]['Value']
```

```python
data = list(groups_dic.values())
labels = list(groups_dic.keys())
plt.figure(figsize = (10,10))
plt.boxplot(data, labels = labels, showfliers=False)
plt.xticks(rotation='vertical')
plt.ylabel('Education Expenditure (%of GDP)')
plt.show()

```
